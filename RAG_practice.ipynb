```python
import pandas as pd
df = pd.read_csv('/content/rotten_tomatoes_top_movies.csv')
df = df[df['title'].notna()] # remove any NaN values as it blows up serialization
df = df[df['consensus'].notna()]
df = df[df['critic_score'].notna()]
df = df[df['genre'].notna()]

df = df.drop(columns=['total_reviews'])
df = df.drop(columns=['crew'])
df = df.drop(columns=['aspect_ratio'])
df = df.drop(columns=['sound_mix'])
df = df.drop(columns=['production_co'])
df = df.drop(columns=['release_date_(theaters)'])
df = df.drop(columns=['director'])
df = df.drop(columns=['producer'])
df = df.drop(columns=['writer'])
df = df.drop(columns=['release_date_(streaming)'])

# data = df.to_dict('records')
data = df.sample(700).to_dict('records')
df
```

```python
from qdrant_client import models, QdrantClient
from sentence_transformers import SentenceTransformer
```

```python
encoder = SentenceTransformer('all-MiniLM-L6-v2') # Model to create embeddings
```

```python
# create the vector database client
qdrant = QdrantClient(":memory:") # Create in-memory Qdrant instance
```

```python
# Create collection to store books
qdrant.recreate_collection(
    collection_name="movie_recommendations",
    vectors_config=models.VectorParams(
        size=encoder.get_sentence_embedding_dimension(), # Vector size is defined by used model
        distance=models.Distance.COSINE
    )
)
```

```python
print(df.columns)
```

```python
# vectorize!
# Note that for Coursera we use an older way of Qdrant doing the uploads using Records instead of Points
qdrant.upload_records(
    collection_name="movie_recommendations",
    records=[
        models.Record(
            id=idx,
            # vector=encoder.encode(f"{doc['consensus']} Critic Score: {doc['critic_score']}, Genre: {doc['genre']}").tolist(),
            vector=encoder.encode(f"{doc['title']} {doc['genre']} {doc['synopsis']} {doc['consensus']}").tolist(),
            payload=doc
        ) for idx, doc in enumerate(data) # data is the variable holding all the wines
    ]
)
```

```python
user_prompt = "Suggest me a mystery movie released in the early 2000s."
```

```python
# Search time!

hits = qdrant.search(
    collection_name="movie_recommendations",
    query_vector=encoder.encode(user_prompt).tolist(),
    limit=6
)
for hit in hits:
  print(hit.payload, "score:", hit.score)
```

```python
# holding the search results
search_results = []
for hit in hits:
  search_results.append(hit.payload)
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

prompt = """
You are a chatbot, a movie specialist. Your priority is to give good movie recommendations based on user requests.

User: Suggest me an amazing horror movie from the early 2000s.
Assistant: Based on your request, here are some suggestions:
"""

for movie in search_results:
    prompt += f"\n- {movie['title']} ({movie['year']}) - {movie['genre']}: {movie['synopsis']} (Critic Score: {movie['critic_score']})"

# Generate a response
response = generator(prompt, max_length=500)
print(response[0]['generated_text'])
```
